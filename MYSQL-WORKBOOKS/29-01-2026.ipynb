```sql
use CG_TRAINING
```

```sql
create table dept(deptid int primary key, deptName varchar(40)) 
```

```sql
insert into dept values(10,'HR');
insert into dept values (20,'Sales')
insert into dept values (30,'Marketing');
```

```sql
create table emp(
    empID int primary key, 
    empname varchar(40),
    worksin int foreign key references dept(deptid)
    )
```

```sql
insert into emp values(101, 'Caleb', 30);
```

```sql
insert into emp values (102, 'Sadie', 10), (103, 'Mike', 20);
```

```sql
insert into emp values (104, 'Eleven', null)
```

```sql
select * from emp
```

```sql
select 
    e.empID, e.empname, d.deptName 
from emp e 
join dept d 
    on e.worksin = d.deptid;
```

```sql
create table doctor (
    docID int primary key,
    docName varchar(50) not null
)

create table patient (
    patID int primary key,
    patName varchar(50) not null,
    docID int,

    CONSTRAINT doc_pat_fk foreign key (docID) references doctor(docID)
)

create table treatment(
    patID int not null,
    docID int not null,
    details varchar(50),

    constraint pat_fk foreign key (patID) references patient(patID),
    constraint doc_fk foreign key (docID) references doctor(docID)
)
```
